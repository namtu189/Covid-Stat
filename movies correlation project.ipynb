```python
# Impoer Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create

# Read in the data
df = pd.read_csv(r'C:\Users\sophi\Downloads\movies.csv')


              
```

```python
#Let look at the data
df.head()
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}.{}%'.format(col, pct_missing))
```

```python
df = df.dropna()
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}.{}%'.format(col, pct_missing))
```

```python
# Data types for our columns
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)

```

```python
pd.set_option('display.max_rows',None)
```

```python
#check for duplicates, drop
df['company'].drop_duplicates.sort_values(ascending=False)
```

```python
# Scatter plot w budget vs gross
plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.scatter(x=df['budget'],y=df['gross'])
```

```python
df.head()
```

```python
# Plot budget vs grooss using seaborn


sns.regplot(x ='budget', y ='gross', data = df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#Let's start looking at correlation - only on numerical

df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
# High correlation btwn budget and gross

correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')
plt.show()
```

```python
#Looks at Company
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerize.head()
```

```python
# High correlation btwn company and gross
sns.regplot(x ='company', y ='gross', data = df_numerized, scatter_kws={"color":"red"}, line_kws={"color":"blue"})

correlation_matrix_company = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix_company, annot=True)
plt.title('Correlation Matric for Numeric features')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')
plt.show()
```
